```python
#Common Libraries
import numpy as np
import pandas as pd
```

```python
#Visualizing my results
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Renaming columns and uploading the data
column_names = ['user_id','item_id','rating','timestamp']
df = pd.read_csv('u.data', sep='\t', names = column_names)
```

```python
#Getting to know the data
#looking at the top 5 rows
df.head()
```

```python
df.describe()
```

```python
#Looking at the data types of each column
df.info()
```

```python
#Now getting the movie titles
movie_titles = pd.read_csv('Movie_Id_Titles')
movie_titles.head()
```

```python
df = pd.merge(df,movie_titles,on='item_id')
df.head()
```

```python
sns.set_style('white')
```

```python
#Performing a groupby to get the average rating for each film
df.groupby('title')['rating'].mean().sort_values(ascending=False).head()
```

```python
#Looking for films with the highest scores
df.groupby('title')['rating'].count().sort_values(ascending=False).head()
```

```python
#Adding the column rating to illustrate the average rating for each film
ratings = pd.DataFrame(df.groupby('title')['rating'].mean())
ratings.head()
```

```python
#Adding a column that indicates the number for ratings for each films
ratings['num of ratings'] = pd.DataFrame(df.groupby('title')['rating'].count())
ratings.head()
```

```python
#A histogram showing the frequency for the number of ratings
plt.figure(figsize=(10,4))
ratings['num of ratings'].hist(bins=70)
```

```python
#A histogram showing the frequency of average scores
plt.figure(figsize=(10,4))
ratings['rating'].hist(bins=70)
```

```python
#performing a pivot to show what rating a user gave to the films they watched
moviemat = df.pivot_table(index='user_id',columns='title',values='rating')
moviemat.head()
```

```python
ratings.sort_values('num of ratings',ascending=False).head(10)
```

```python
starwars_user_ratings = moviemat['Star Wars (1977)']
liarliar_user_ratings = moviemat['Liar Liar (1997)']
starwars_user_ratings.head()
```

```python
similar_to_starwars = moviemat.corrwith(starwars_user_ratings)
similar_to_liarliar = moviemat.corrwith(liarliar_user_ratings)
```

```python
#Removing the rows that contain NaNs
#
corr_starwars = pd.DataFrame(similar_to_starwars,columns=['Correlation'])
corr_starwars.dropna(inplace=True)
corr_starwars.head()
```

```python
corr_starwars.sort_values('Correlation',ascending=False).head(10)
```

```python
corr_starwars = corr_starwars.join(ratings['num of ratings'])
corr_starwars.head()
```

```python
corr_starwars[corr_starwars['num of ratings']>100].sort_values('Correlation',ascending=False).head()
```

```python
corr_liarliar = pd.DataFrame(similar_to_liarliar,columns=['Correlation'])
corr_liarliar.dropna(inplace=True)
corr_liarliar = corr_liarliar.join(ratings['num of ratings'])
corr_liarliar[corr_liarliar['num of ratings']>100].sort_values('Correlation',ascending=False).head()
```
